```python
import pandas as pd
import numpy as np

train_file =pd.read_csv("1601355459_1343217_train.csv")
test_file =pd.read_csv("1601355459_1391656_test.csv")



```

```python
#preprocessing 
train_file["F10"] = train_file["F10"].str.lower()
train_file["F11"] = train_file["F11"].str.lower()

test_file["F10"] = test_file["F10"].str.lower()
test_file["F11"] = test_file["F11"].str.lower()

```

```python
#changing object type to string
train_file["F11"] = train_file.F11.astype('string')
test_file["F11"] = test_file.F11.astype('string')


```

```python
# change the string type to number since it will be easy to compare data later while using models

#changing F11 to 1 and 0. 1 represent male and 0 represent female
# for train_file
change_F11_forTrainFile = pd.get_dummies(train_file['F11'])
var_F11ForTrainFile = pd.DataFrame(change_F11_forTrainFile)
var1 = var_F11ForTrainFile.drop(var_F11ForTrainFile.columns[0],axis=1)


#for test_file
change_F11_forTestFile = pd.get_dummies(test_file['F11'])
var_F11ForTestFile = pd.DataFrame(change_F11_forTestFile)
var2 = var_F11ForTestFile.drop(var_F11ForTestFile.columns[0],axis=1)
```

```python
#Now dropping F11 from the dataset and renamining female to F11. This way I remove string type and replace that with 1's and 0's
train_file = train_file.drop(['F11'], axis=1)
test_file = test_file.drop(['F11'], axis=1)

```

```python
var1 = var1.rename(columns={" male": "F11"})
var2 = var2.rename(columns={" male": "F11"})

concat_onTrainFile = pd.concat((train_file,var1),axis=1)
concat_onTestFile = pd.concat((test_file,var2),axis=1)
```

```python
## changing F10 string to int values 
#for Train_file
conditions = [
    (concat_onTrainFile['F10'] == " white"),
    (concat_onTrainFile['F10'] == " black"),
    (concat_onTrainFile['F10'] == " asian-pac-islander"),
    (concat_onTrainFile['F10'] == " amer-indian-eskimo"),
    (concat_onTrainFile['F10'] == " other")
    ]

# Values assinging unique numbers
values = ['4', '3', '2', '1','0']
concat_onTrainFile['F10_data'] = np.select(conditions, values)


#for Test_file
conditions_forTestFile =  [
    (concat_onTestFile['F10'] == " white"),
    (concat_onTestFile['F10'] == " black"),
    (concat_onTestFile['F10'] == " asian-pac-islander"),
    (concat_onTestFile['F10'] == " amer-indian-eskimo"),
    (concat_onTestFile['F10'] == " other")
    ]

# Values assinging unique numbers
values_forTestFile = ['4', '3', '2', '1','0']
concat_onTestFile['F10_data_TestFile'] = np.select(conditions_forTestFile, values_forTestFile)
```

```python
#drop F10 object row
concat_onTrainFile = concat_onTrainFile.drop(['F10'], axis = 1)
concat_onTestFile =  concat_onTestFile.drop(['F10'], axis = 1)
```

```python
#renaming column
concat_onTrainFile = concat_onTrainFile.rename(columns={"F10_data": "F10"})
concat_onTestFile = concat_onTestFile.rename(columns = {"F10_data_TestFile": "F10"})
```

```python
#rearrangeing columns 
#For train_file
rearrange_colums_TrainFile = ["F1","F2","F3","F4","F5","F6","F7","F8","F9","F10","F11","credit"]
concat_onTrainFile = concat_onTrainFile.reindex(columns=rearrange_colums_TrainFile)

#For Test_file
rearrange_colums_TestFile = ["F1","F2","F3","F4","F5","F6","F7","F8","F9","F10","F11","credit"]
concat_onTestFile = concat_onTestFile.reindex(columns=rearrange_colums_TestFile)
concat_onTestFile =  concat_onTestFile.drop(['credit'], axis = 1)
```

```python
concat_onTrainFile["F10"] = pd.to_numeric(concat_onTrainFile["F10"])
concat_onTestFile["F10"] = pd.to_numeric(concat_onTestFile["F10"])
```
